1번 문제
========
1. num_1 폴더의 uriage 파일을 pandas를 이용하여 데이터프레임으로 불러온다. 
2. purchase_date 컬럼의 데이터 타입을 datetime을 변경한다. 
3. item_name 컬럼의 값을 대문자로 변경한다. 
4. item_name 컬럼의 공백을 지워준다. 
5. item_name 컬럼을 기준으로 오름차순 정렬한다. 
6. 데이터의 결측치가 존재하는지 확인한다. 
7. item_price 컬럼의 결측치의 값을 해당하는 item_name의 맥스값으로 치환하여 준다. 

```python
import pandas as pd
uriage = pd.read_csv('./num_1/uriage.csv')
uriage['purchase_date']=pd.to_datetime(uriage['purchase_date'])
uriage['item_name']=uriage['item_name'].str.upper()
uriage['item_name']=uriage['item_name'].apply(lambda a: a.replace(' ',''))
uriage['item_name'].sort_values(ascending=False,)
uriage.isna().sum()
uriage['item_price'].fillna(2600,inplace=True)

uriage
```
