```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

interest_rate_df = pd.read_csv('./예금금리비교.csv',encoding='cp949')
interest_rate_df
```

```python
interest_rate_df2 = interest_rate_df.drop(['Unnamed: 4','Unnamed: 5','Unnamed: 8','Unnamed: 9','기본금리(단리이자 %)','최고우대금리(단리이자 %)',
                                            '은행 최종제공일','만기 후 금리'],axis=1)
interest_rate_df2.rename(columns={'Unnamed: 3':'기본금리','Unnamed: 7':'최고우대금리'},inplace=True)
interest_rate_df3 = interest_rate_df2.drop(columns=['가입방법','우대조건','가입 제한조건','가입대상','기타 유의사항','최고한도'],axis=1)
interest_rate_df3.drop(index=0,axis=0,inplace=True)


basic_h10_df = interest_rate_df3.sort_values('기본금리',ascending=False).head(10)
vip_h10_df = interest_rate_df3.sort_values('최고우대금리',ascending=False).head(10)

basic_l10_df = interest_rate_df3.sort_values('기본금리',ascending=True).head(10)
vip_l10_df = interest_rate_df3.sort_values('최고우대금리',ascending=True).head(10)

basic_h10_df.reset_index(drop=True,inplace=True)
vip_h10_df.reset_index(drop=True,inplace=True)
basic_l10_df.reset_index(drop=True,inplace=True)
vip_l10_df.reset_index(drop=True,inplace=True)


interest_rate_df3.to_csv('./은행금리/은행금리(수정).csv', encoding='cp949')
basic_h10_df.to_csv('./은행금리/기본금리_상위10개소.csv', encoding='cp949')
basic_l10_df.to_csv('./은행금리/기본금리_하위10개소.csv', encoding='cp949')
vip_h10_df.to_csv('./은행금리/우대금리_상위10개소.csv', encoding='cp949')
vip_l10_df.to_csv('./은행금리/우대금리_하위10개소.csv', encoding='cp949')
```

```python
basic_h10_interest = basic_h10_df['기본금리'].tolist()
basic_l10_interest = basic_l10_df['기본금리'].tolist()
st_interest = [1.25,1.25,1.25,1.25,1.25,1.25,1.25,1.25,1.25,1.25]
index = np.arange(1,11,1)
print(index)
print(basic_l10_interest)

fig, ax = plt.subplots()
ax.plot(index,basic_h10_interest)
ax.plot(index,basic_l10_interest)
ax.plot(index,st_interest)
ax.set(xlim=(0,10),xticks=index, ylim=(0,5),yticks=(0,15))
plt.show()
```
