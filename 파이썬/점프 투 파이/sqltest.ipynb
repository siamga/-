```python
sql_2 = '''
        SELECT Country, SUM(`Units Sold`) as sum 
        FROM sales GROUP BY Country
        ORDER BY Country
        '''
cursor.execute(sql_2)
result=cursor.fetchall()
pd.DataFrame(result)
```

```python
sql = '''
      SELECT Country, COUNT(`Country`) as cnt,
             SUM(`Units Sold`) as sum,
             CHAR_LENGTH(Country) as length
      FROM sales
      GROUP BY Country
      ORDER BY char_length(`Country`) desc
      '''
cursor.execute(sql)
result=cursor.fetchall()
pd.DataFrame(result)
```

```python
sql = '''
      SELECT DISTINCT Region from sales
      '''
cursor.execute(sql)
result=cursor.fetchall()
pd.DataFrame(result)
```

```python
sql = '''
      SELECT 
      Country,
      IF(CHAR_LENGTH(Country)>5,'LONG','SHORT') AS length
      from sales
      Order by char_length(Country) desc
      '''
cursor.execute(sql)
result = cursor.fetchall()
pd.DataFrame(result)
```

```python
sql = '''
      SELECT 
      Region,
      Country,
      ISNULL('Item Type') as item
      from sales
      '''
cursor.execute(sql)
result = cursor.fetchall()
pd.DataFrame(result)
```
