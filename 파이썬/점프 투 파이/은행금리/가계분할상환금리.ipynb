```python
#은행 분할 대출 금리 일람
import pandas as pd
import numpy as np

#분할상환 방식
loan_dev = pd.read_csv('.\대출금리\가계대출\가계대출금리(분할상환_주담대).csv', encoding='cp949')

loan_dev.drop([0,6,15,17],axis=0,inplace=True)
loan_dev = loan_dev.apply(lambda x : x.replace('-',0))
loan_dev['Unnamed: 7'] = pd.to_numeric(loan_dev['Unnamed: 7'])
#loan_dev_bank = loan_dev['은행'].values.tolist()
#loan_dev_avg = loan_dev['Unnamed: 7']
#loan_dev2 = pd.DataFrame(loan_dev_avg,index = loan_dev_bank)

#loan_dev2.columns = ['평균금리']
loan_dev.columns=['은행','구분','1~2등급','3~4등급','5~6등급','7~8등급','9~10등급','평균금리','참고사항']
loan_dev['은행'] = loan_dev['은행'].str.strip()
#loan_dev = loan_dev.drop(['Unnamed: 3','Unnamed: 4','Unnamed: 5','Unnamed: 6','구분',
#                           '참고사항','신용등급별 금리(%)'],axis=1)
loan_dev.reset_index(drop=True,inplace=True)
loan_dev
```

```python
credit_distribution = pd.read_csv('./신용등급분포.csv', encoding='cp949')
cd = np.array(credit_distribution['구성 비율'],dtype='float')
ac_cd = np.array(credit_distribution['누적비율'],dtype='float')


interest_grade = loan_dev[['1~2등급','3~4등급','5~6등급','7~8등급','9~10등급']]
credit_distribution
interest_grade

```

```python
#이거 분명히 함수로 만들어서 하는 방법이 있는거 같은데 못찾겠다.......
credit_distribution['case1'] = [1,0,0,0,0]

case1_1 = credit_distribution['구성 비율'][0]/credit_distribution['누적비율'][1]
case1_2 = credit_distribution['구성 비율'][1]/credit_distribution['누적비율'][1]
credit_distribution['case2'] = [case1_1,case1_2,0,0,0]

case2_1 = credit_distribution['구성 비율'][0]/credit_distribution['누적비율'][2]
case2_2 = credit_distribution['구성 비율'][1]/credit_distribution['누적비율'][2]
case2_3 = credit_distribution['구성 비율'][2]/credit_distribution['누적비율'][2]
credit_distribution['case3'] = [case2_1,case2_2,case2_3,0,0]

case3_1 = credit_distribution['구성 비율'][0]/credit_distribution['누적비율'][3]
case3_2 = credit_distribution['구성 비율'][1]/credit_distribution['누적비율'][3]
case3_3 = credit_distribution['구성 비율'][2]/credit_distribution['누적비율'][3]
case3_4 = credit_distribution['구성 비율'][3]/credit_distribution['누적비율'][3]
credit_distribution['case4'] = [case3_1,case3_2,case3_3,case3_4,0]

credit_distribution['case5'] = credit_distribution['구성 비율']
credit_distribution
```

```python
cd = np.array(credit_distribution['구성 비율'],dtype='float')
ac_cd = np.array(credit_distribution['누적비율'],dtype='float')

interest_grade = loan_dev[['1~2등급','3~4등급','5~6등급','7~8등급','9~10등급']]


result = np.array(interest_grade.loc[0:0])


def multi(x):
     #rate_grade=np.array(interest_grade[:][x:x+1],dtype='float')[0]
     rate_grade=np.array(interest_grade.loc[x:x,'1~2등급':'9~10등급'], dtype='float')[0]
     
     if rate_grade[0]==0:
         return rate_grade@np.zeros((5,1))
     elif rate_grade[1] == 0:
        return rate_grade@credit_distribution['case1']
     elif rate_grade[2] == 0:
        return rate_grade@credit_distribution['case2']
     elif rate_grade[3] == 0:
        return rate_grade@credit_distribution['case3']
     elif rate_grade[4] == 0:
        return rate_grade@credit_distribution['case4']
     else:
        return rate_grade@credit_distribution['case5']


a=[]
for i in range(len(interest_grade)):
    a.append(multi(i))

interest_grade['가중평균수익'] = a

interest_grade


#interest_grade[:][15:16]
#credit_distribution['case1(1등급 대상 고객분포)'] #이게 왜 안돼!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
```
