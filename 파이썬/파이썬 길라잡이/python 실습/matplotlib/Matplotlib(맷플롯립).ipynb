```python
#맷플롯립 : 데이터 시각화 패키지
#cmd에서 설치: pip install matplotlib

# plot() : 하나의 숫자로 이뤄진 리스트를 그래프로 지정한다.
# show() : 그래프를 화면에 출력한다.

import matplotlib.pyplot as plt
plt.plot([1,2,3,4])
plt.show()
```

```python
# plot()에 두개의 리스트를 넣으면 첫번째가 x축, 두번째가 y축 값이 된다.
plt.plot([1,2,3,4],[1,4,5,10])
plt.show()
```

```python
# 선색 변경: plot(x리스트, y리스트, 옵션)
# 옵션 부분에 선색의 변경이 가능하다.(디폴트:b(파랑))
#r(빨강), g(초록), c(하늘색), m(마젠타)
#y(노랑), k(검정), w(하얀색) 혹은 색상표의 코드를 입력 가능
plt.plot([1,2,3,4],[1,4,5,10],'r')
plt.show()


##옵션에 선 종류를 입력하면 선 종류가 바뀐다.
##-(단선), --(끊긴 줄), -.(줄점줄)

'''
옵션의 수가 많아지면 그냥 ''안에 다 넣으면 된다.
'''
plt.plot([1,2,3,4],[1,4,5,10],'r--D')
plt.plot([1,2,3,4],[5,3,7,1],'ob')
plt.plot([1,2,3,4],[8,2,5,3],'.m')
plt.show()
```

```python
#plt.legend() : 그래프 라인의 범례 추가
#plt.xlabel(), plt.ylabel() -> x,y축 레이블 추가
#plt.legend()에서 loc='위치'를 지정하면 해당 위치에 
# 범례가 나온다.
#plt.title() -> 그래프 제목 추가

plt.plot(['maths','science','history'],[90,95,60])
plt.plot(['maths','science','history'],[65,70,95])
plt.xlabel('subject')
plt.ylabel('score')
plt.title('Score Board')
plt.legend(['A','B'])
plt.show()

plt.plot(['maths','science','history'],[90,95,60])
plt.plot(['maths','science','history'],[65,70,95])
plt.xlabel('subject')
plt.ylabel('score')
plt.title('Score Board')
plt.legend(['A','B'],loc = 'upper right')
plt.show()

#loc = '위치'를 지정하면 아래와 같이 지정하면 해당 위치에 범례 등장
#상하위치 좌우 위치
#upper     left 
#center   center  -> 정중앙은 그냥 center
#lower     right
```

```python
#서브 플롯: 여러개의 그래프를 한 공간에 같이 그릴 때 사용.
#그래프의 위치를 격자형으로 지정해서 그래프를 출력한다.
#plt.subplot(nrow,ncol,pos)
#nrow=행의 수, ncol=열의 수, pos=위치
plt.subplot(2,1,1)
plt.plot([3,5,1,7])
plt.subplot(2,1,2)
plt.plot([1,2,3,4])
plt.show()

plt.subplot(1,2,1)
plt.plot([3,5,1,7])
plt.subplot(1,2,2)
plt.plot([1,2,3,4])
plt.show()
```

```python
#바형 그래프: bar()함수를 사용하여 출력한다.
x = [1,2,3]
y = [80,90,65]

plt.bar(x,y)
plt.show()

#xticks(): x축의 레이블 값의 수정이 가능하다. 
plt.bar(x,y)
plt.xticks(x,['math','science','history'])
plt.show()

#바형 색상 변경
#바형은 옵션의 숫자가 많아서 color=로 지정을 해야 
#그래프 색상이 바뀐다. 이때 색이 리스트로 제시될 때, 
#전체 바의 개수보다 작다면 반복되서 색깔이 바뀐다.

plt.bar(x,y,color=['r','g'])
plt.xticks(x,['math','science','history'])
plt.show()

#바형 폭 변경: width= 로 변경한다.
plt.bar(x,y,color=['r','g'],width=0.1)
plt.xticks(x,['math','science','history'])
plt.show()
```

```python
#산점도의 표현: scatter()를 사용
import numpy as np

x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x,y)
plt.show()

#크기 및 색상 지정
area = (20 * np.random.rand(50))**2
colors = np.random.rand(50)

plt.scatter(x,y,s=area,c=colors)
plt.show()
```
