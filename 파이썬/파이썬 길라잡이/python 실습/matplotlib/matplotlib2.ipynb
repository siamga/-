```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(10,100,10)
x
```

```python
#벡터에서 짝수 혹은 홀수 번째 원소를 출력한다
x_even = np.array(x[1:9:2])
x_odd = np.array(x[0:9:2])
print(x_even)
print(x_odd)

# x[시작:종료:증가값]
```

```python
#np.linspace(시작값,종료값,분할 수): 시작값과 종료값을 포함해서 전채를 분할 수만큼 나눠서 출력한다.
x = np.linspace(0,10,4)
plt.plot(x)
plt.show()
```

```python
#np.logspace(시작,종료,분할수): 시작값*log10(10)에서 종료값*log10(10)을 분할 수 만큼 나눠서 출력한다.
x = np.logspace(2,5,50)
plt.plot(x)
plt.show()
```

```python
#np.random.seed(seed값): 특정 seed값으로 지정한 랜덤 값을 출력한다.
#만일 seed값이 같은 random 결과가 여러개 있다면 그 값은 동일하다.
# np.random.randint(시작,종료,n): 시작과 종료-1 까지의 숫자 중에서 n개의 정수를 출력한다.

np.random.seed(1)
result1 = np.random.randint(10,100,10)

np.random.seed(2)
result2 = np.random.randint(10,100,10)

np.random.seed(3)
result3 = np.random.randint(10,100,10)

result1, result2, result3
```

```python
#np.random.rand(n): n개의 원소를 가진 벡터를 0~1 사이의 실수에 대해서 같은 확률로 뽑아서 출력한다.
np.random.rand(10)
```

```python
#np.random.randn(n): 표준 정규 분포에서 n개의 실수를 출력한다.
np.random.randn(10)
```

```python
#np.random.randint(): n 부분에 행렬 모양을 넣어서 행렬을 출력할 수 있다.
r = np.random.randint(1,100,(3,4))
r
```

```python
#행렬 내의 행의 위치 변경: np.random.shuffle(행렬)
np.random.shuffle(r)
r
```

```python
#제곱근: np.sqrt()-> 스칼라를 넣으면 스칼라의 제곱근, 행렬이나 벡터를 넣으면 원소들의 제곱근이 출력된다.
a = [1,9,25,49]
a_sqrt = np.sqrt(a)
a_sqrt
```

```python
#제곱: np.sq()-> 스칼라를 넣으면 스칼라의 제곱, 행렬이나 벡터를 넣으면 원소들의 제곱이 출력된다.
a = [1,3,5,7]
a_sq = np.square(a)
a_sq
```

```python
#np.exp():자연상수(e)가 밑인 지수 함수의 결과값 출력(f(x)=e^x)
np.exp(0)
np.exp(4)
```

```python
x = np.linspace(-5,5,101)
x
```

```python
y = (1 / np.sqrt(2*np.pi))*np.exp(-x**2/2)
y
```

```python
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import scipy.stats as stats
```

```python
# scipy.stats.norm(avg,std): 평균이 avg이고 표준편차가 std인 정규분포의 값을 출력한다.
# pdf:확률 밀도 함수
# pmf: 확률 질량 함수
# cdf: 누적 분포 함수


y_ = stats.norm(0,1).pdf(x)
y_
plt.plot(x,y_)
plt.show()
```

```python
def norm(a,b,_x):
    return stats.norm(a,b).pdf(_x)

x = np.linspace(-20,20,101)
x
```

```python
for i in range(1,11, 1):
    _y = norm(0,i,x)
    plt.plot(x,_y)

plt.show()
    
```

```python
# histogram

x = np.random.randint(0,10,10)
```

```python
#plt.hist(s,bins=10): 특정 구간에 있는 데이터의 개수를 시각화하는(=히스토그램 작성) 함수
#            기본적으로 10개의 구간이 디폴트 값이다.
plt.hist(x)
plt.hist(x, bins=30)
plt.show()
```

```python
x = np.random.randint(10,50,5)
labels = ['A','B','C','D','E']
plt.pie(x, autopct='%.2f%%',labels=labels)
plt.show()
```

```python
plt.pie(x, autopct='%.2f%%',labels=labels, counterclock=False)
plt.show()
```

```python
plt.pie(x, autopct='%.2f%%',labels=labels, 
counterclock=False, startangle=180)
plt.show()
```

```python
explodes = [0,0.1,0.1,0,0.2]
plt.pie(x, autopct='%.2f%%',labels=labels, 
counterclock=False, startangle=180,
explode = explodes)
plt.show()
```

```python
x = np.linspace(1,10,10)
y = np.random.rand(10)
```

```python
#x,y 축의 범위 지정: plt.xticks(구간의 눈금 리스트)-> x축 범위 리스트
#                   plt.yticks(구간의 눈금 리스트)-> y축 범위 리스트
plt.plot(np.around(x,2),np.around(y,2))
plt.xticks(np.around(np.arange(min(x),max(x)+1,1),2))
plt.xlabel('x')

plt.yticks(np.arange(min(y),max(y)+0.1,0.1))
plt.ylabel('y')

plt.legend('x graph')
plt.grid()

plt.show()
```
