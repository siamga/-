```python
#
sql_16 = '''
            SELECT 
            ENAME,
            CASE WHEN JOB = 'ANALYST' THEN '분석가'
                WHEN JOB = 'CLERK' THEN '점원'
                WHEN JOB = 'MANAGER' THEN '관리자'
                WHEN JOB = 'PRESIDENT' THEN '사장'
                WHEN JOB = 'SALESMAN' THEN '판매사원'
                END AS JOB
            FROM emp
         '''
result_16 = _db.executeAll(sql_16)
df_16 = pd.DataFrame(result_16)
df_16
```

1. emp table과 dept table을 left join 조건 dept가 같은 경우
2. 출력 컬럼은 ENAME,JOB,SAL,DEPTNO,LOC
3. ENAME, LOC 그대로 출력
4. JOB은 각각 한글명 변경
5. SAL은 2450 기준 이상이면 HIGH 낮으면 LOW
6. DEPNO은 DEPTNO의 값과 dept table의 Dname 값을 더해서 표시 (DEPTNO-DNAME 형식)
7. 모든 컬럼 이름은 한글로 변경

```python
sql_t = '''
        SELECT 
        emp.ENAME AS '사원명',
        CASE WHEN emp.JOB='SALESMAN' THEN '영업사원'
             WHEN emp.JOB='CLERK' THEN '점원'
             WHEN emp.JOB='MANAGER' THEN '관리자'
             WHEN emp.JOB='ANALYST' THEN '분석가'
             WHEN emp.JOB='PRESIDENT' THEN '사장'
             END AS '직종',
        IF(emp.SAL >= 2450,'HIGH','LOW') AS '월급',
        CONCAT(emp.DEPTNO,'-',dept.DNAME) AS '부서',
        dept.LOC AS '위치'
        FROM emp left join dept 
        ON emp.DEPTNO = dept.DEPTNO
        ORDER BY `사원명` ASC
        '''
#JOIN에서 데이터 작성 시, 테이블명이 먼저 나온다!
result_t = _db.executeAll(sql_t)
df_t = pd.DataFrame(result_t)
df_t
```

```python
sql_t_2 = '''SELECT emp.ENAME,emp.JOB,emp.SAL,emp.DEPTNO,dept.DNAME,dept.LOC from emp left join dept on emp.DEPTNO = dept.DEPTNO'''
result_t_2 = _db.executeAll(sql_t_2)
df_t_2 = pd.DataFrame(result_t_2)
df_t_2.columns=['사원명','직종','급여','부서번호','부서명','위치']
df_t_2['직종']=df_t_2['직종'].apply(lambda x:'점원' if x=='CLERK' else('영업사원' if (x=='SALESMAN') else('관리자' if (x=='MANAGER') else('분석가' if (x=='ANALYST') else '사장' ))))
df_t_2['급여']=df_t_2['급여'].apply(lambda x: 'High' if (x >= 2500) else 'Low')

df_t_2
```

```python
sql_19 = '''
            SELECT
            (
                SELECT COUNT(*) FROM emp
            ) AS `CNT(emp)`,
            (
                SELECT COUNT(*) FROM dept
            ) AS `CNT(dept)`
         '''
result_19 = _db.executeAll(sql_19)
pd.DataFrame(result_19)
```

1.emp: 사원 정보 테이블 
dept: 부서 정보 테이블

```python
sql_20 = '''
         SELECT emp.ENAME
         FROM emp LEFT JOIN dept 
         ON emp.DEPTNO = dept.DEPTNO
         WHERE dept.DNAME = 'sales'
         '''

result_20 = _db.executeAll(sql_20)
pd.DataFrame (result_20)
```

```python
#부서 위치가 'NEW YORK','DALLAS'가 아닌 소속의 사원 이름을 출력하라
sql_22 = ''' 
         SELECT 
         ENAME
         FROM emp
         WHERE DEPTNO IN  (
                             SELECT DEPTNO
                             FROM dept
                             WHERE LOC NOT IN ('NEW YORK','DALLAS')
         )
         '''

result_22 = _db.executeAll(sql_22)
pd.DataFrame(result_22)
```

```python
#데이터의 일부가 같은 값들을 출력하고자 할 때, LIKE '부분 데이터%' :부분 데이터가 맨 앞에 있는 데이터 추출(in python:부분데이터%%)
#                                                 '%부분데이터' :부분 데이터가 맨 뒤에 있는 데이터 추출(in python: %%부분데이터)
#                                                  '%부분데이터%' :부분 데이터가 있는 데이터 추출(in python: %%부분데이터%%)
#단, 파이썬에서 사용하려면 %를 %%로 바꿔서 쓴다.

#특정 위치에 특정 값이 있는 데이터를 뽑으려면 _(언더바)를 이용해서 위치를 표시하고 %%를 사용한다.
sql_25 = '''
         SELECT * FROM emp
         WHERE ENAME LIKE '%%s'
         '''
result_25 = _db.executeAll(sql_25)
pd.DataFrame(result_25)

sql_26 = '''
         SELECT * FROM emp
         WHERE ENAME LIKE '%%E_'
         '''
result_26 = _db.executeAll(sql_26)
pd.DataFrame(result_26)
```
