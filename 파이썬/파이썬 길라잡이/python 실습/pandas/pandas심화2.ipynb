```python
import pandas as pd 

corona_df = pd.read_csv('./corona.csv')
corona_df
```

```python

corona_df.columns = ['인덱스','등록일시','누적사망자','누적확진자','게시글번호','기준일','기준시간','수정일시','누적의심자','누적확진률']
#.describe()-> 데이터프레임의 정보 요약: cuont->columns별 데이터 개수 
#                                       mean,std -> 평균, 표준편차
#                                       min,25%,50%,75%,max -> 최소, 분위값, 최대값

corona_df.describe()
```

```python
#특정 열에 대한 정보 요약
corona_df["누적확진률"].describe()
```

```python
#시간에 따른 데이터 정렬
corona_df_2 = corona_df.sort_values('등록일시')
corona_df.sort_values('등록일시')
```

```python
#데이터의 수정
#.shift(n)-> n번 아래의 데이터 선택(디폴트=1)
#.diff() -> 바로 직전 데이터와의 차이
corona_df_2['일일확진자'] = (corona_df_2['누적확진자']-corona_df_2['누적확진자'].shift()).fillna(0)
corona_df_2['일일사망자'] = corona_df_2['누적사망자'].diff().fillna(0)
corona_df_2
```

```python
#.value_counts(): 같은 value 값의 등장 개수
corona_df_2['일일사망자'].value_counts(ascending=True) #여기서의 ascending은 count의 개수를 기준으로 정렬 된다.
```

```python
# 데이터 삭제
corona_df_3 = corona_df_2.drop(['인덱스','기준시간','게시글번호','기준일','수정일시'], axis=1)
corona_df_3
```

```python
#.reset_index() 활용: 인덱스를 초기화해서 새로 만들어 놓는다.(디폴트:drop=False, inplace=False -> 기존의 인데스를 놔둔다. 원본 데이터는 그대로 둔다.)
corona_df_3.reset_index()

#.drop= : 기존 인덱스 삭제 여부 옵션(True=삭제한다, False=놔둔다)
#inplace= : 원본 데이터에 대한 수정 여부 옵션(True=원본 데이터에 적용, False=원본데이터 미적용)

corona_df_3.reset_index(drop=True)
```

```python
corona_df_4 = corona_df_3.tail(150)
corona_df_4
```

```python
#fillna(method='ffill'이나 'bfill'): NaN으로 지정된 칸에 데이터를 채우는 함수
#ffill 옵션: NaN 값 직전 값을 넣어준다
#bfill 옵션: NaN 값 직후의 값을 넣어준다.

corona_df_4.fillna(corona_df.mean())
```
