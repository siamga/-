```python
import pandas as pd
sales_df = pd.read_csv('Sales Records.csv')
sales_df
```

```python
#1.컬럼 삭제: Ship Date, Order ID, Sales Channel, Total Revenue ->새로운 변수 작성 후 계속 작업
#2.컬럼명 변경: 
#3.Country 기준 sort 
#4.index 값 초기화
sales_df2 = sales_df.drop(['Ship Date','Order ID','Sales Channel','Total Revenue'],axis=1)


sales_df2.columns=['지역','국가','물건','배송순위','주문날짜','판매개수','단가','원가','총비용','총수익']


sales_df2.sort_values('국가',inplace=True)

sales_df2.reset_index(drop=True, inplace=True)
sales_df2
```

```python
#.astype
sales_df2['총수익'] = sales_df2['총수익'].astype(int)
sales_df2
```

```python
#round(n): 소수점 n번째를 남기고 반올림하는 함수
sales_df2['총비용'] = sales_df2['총비용'].round(0).astype(int)
sales_df2
```

```python
sales_df2['원가'] = sales_df2['원가'].round(1)
sales_df2
```

```python
#round(-n):정수 부분에서 반올림한다.(n은 자연수)
sales_df2['판매개수'].round(-1)
```

```python
sales_df2['주문날짜'] = pd.to_datetime(sales_df2['주문날짜'], 
                            format = '%m/%d/%Y')

sales_df2
```

```python
#.groupby('column의 이름')->특정 columns의 values를 기준으로 그룹화를 한 이후 .함수를 붙여서 그 수치를 보여준다.
sales_df2.groupby(['지역','국가','물건']).sum()['총수익']
sales_df2.groupby('지역').mean()['총수익']
```

```python
#.get_group(): groupby로 묶은 그룹 중에서 특정 그룹의 값만 출력할 때 사용하는 함수이다.
sales_df2[sales_df2['국가'].isin(['South Korea','Turkey'])]
sales_df2['국가'].isin(['South Korea','Turkey'])
```

```python
import numpy as np
s1 = pd.Series([1, np.nan,3,4,5])
s2 = pd.Series([1,2,np.nan,4,5])
s3 = pd.Series([1,2,3,np.nan,5])

df = pd.DataFrame({'s1':s1,
                's2':s2,
                's3':s3})
df
```

```python
#데이터프레임의 연산

#스칼라 연산
df['s1'] = df['s1'] +1
print(df)

#행렬 단순연산
_list = (5,4,3,2,1)
df['s2'] =df['s2'] + _list
print(df)
```
