```python
import pandas as pd

values = [1,2,3,4,5,1,2,3]
test = pd.DataFrame(values)
test
```

```python
test.columns=['A']
test
```

```python
test['B'] = test['A'].apply(lambda j : 0 if (j <= 3) else 1)
test
```

```python
#test['A']가 1,2면 0, 3,4면 1, 
def change(x):
    if x < 3 :
        return 0
    elif x <5:
        return 1
    elif x < 7 :
        return 2
    else:
        return 3

test.A = test.A.apply(change)
test
```
