```python
#class를 생성하라
#1. 인자값으로 url(주소) -> csv 파일의 주소
#2. 함수 1개 생성 -> url에 있는 파일을 읽어오고 'Country'기준으로 sort, 인덱스 값을 초기화하라
#3. 함수 1개 생성 -> 인자값을 list로 받는다.-> 이 매개변수 기준으로 columns를 삭제
import pandas as pd
import numpy as np
class sales():
    def __init__(self,url):
        self.url = url
    
    #함수의 디폴트값 설정하는 법 => 함수의 매개변수를 지정할 때, '=디폴트값'을 붙여놓는다.
    def tidy(self,input_sort='Country',input_inplace=True):
        self.df = pd.read_csv(self.url)
        self.df.sort_values(input_sort,inplace=input_inplace)
        self.df.reset_index(drop=True, inplace=input_inplace)
        return self.df
    
    def columns(self,_list = []):
        self.df.drop(_list,axis=1,inplace=True)
        return self.df
    
    def re_name(self,col):
        self.df.columns = col
        return self.df

    def grouping(self,nation):
        self.total = self.df.groupby('Country').get_group(nation)
        return self.total

    def drop_set(self,start,end):
        self.df.drop(self.df.loc[:,start:end],axis=1,inplace=True)
        return self.df
    

a = sales('Sales Records.csv')
a.tidy()
a.columns()
a.drop_set('Order Date','Unit Cost')
```
