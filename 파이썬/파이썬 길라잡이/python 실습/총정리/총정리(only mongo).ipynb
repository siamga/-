```python
from pymongo import MongoClient
import pandas as pd
```

```python
client = MongoClient(
    host='localhost',
    port=27017
)

client.list_database_names()
```

```python
db = client['ubion']
pd.DataFrame(db.hmm.find())
```

```python
result = db.hmm.find(
    {},
    {
        'Date':1,
        'Open':1,
        'Close':1,
        'High':1,
        'Low':1,
        'Volume':1,
        '_id':0
    }
)
result = pd.DataFrame(result)
```

```python
result['UP_DOWN'] = result['Close']-result['Open']
result['UD_RATION'] = result['UP_DOWN']/result['Open']*100
result = result.round(2)
result
```
