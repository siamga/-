```python
import yfinance as yf
import pandas as pd

hmm = yf.Ticker('011200.ks')
result = hmm.history(period='max')
result
```

```python
result = result.round(2)
result.reset_index(inplace=True,drop=True)
result = result.loc[:999,:]
result
```

```python
result['UP_DOWN'] = result['Open'] - result['Close']
result['UD_RATION'] = result['UP_DOWN']/result['Open'] * 100
result.drop(['Stock Splits','Dividends'],axis=1, inplace=True)
result.round(2)
```

```python
import matplotlib.pyplot as plt

plt.plot(result.Open)
plt.show()
```
