```python
# emp collection에 대해서
# key값이 EMPNO,ENAME,DEPTNO만 출력되도록 코드를 작성하라
result = db.emp.find(
    {},
    {'EMPNO':1,'ENAME':1,'DEPTNO':1,'_id':0}
    )

result2 = db.emp.find(
    {},
    {'EMPNO':1,'ENAME':1,'DEPTNO':1,'_id':0}
    )

pd.DataFrame(result2)

#여기서 몽고db는 '='표시 대신 ':' 표시로 데이터를 지정한다는 점을 
#주의해야한다. 딕셔너리 형태로 조건을 지정해야하니까.
```

```python
##.sort(key,1이나 -1): 어떤 key의 value 값을 기준으로 collection을 
## 오름차순(1),내림차순(-1)으로 정렬하는 함수 디폴트는 1이다
result = db.emp.find().sort('ENAME',-1)

pd.DataFrame(result)

# SELECT * FROM emp ORDER BY ENAME
```

```python
#특정 범위의 데이터 출력
#find() 안에 범위 조건을 적는다.
#범위 조건: $gt : 초과
#          $gte : 이상
#          $lt : 미만
#          $lte : 이하
#          $eq : 등호
#

result = db.emp.find(
    {
        'SAL':{
            '$gte':2500,
            '$lte':3000
        }
    }
)
pd.DataFrame(result)


## SELECT * FROM emp WHERE SAL >= 2450
```

```python
#부분 조건을 만족하는 데이터의 출력
result = db.emp.find(
    {
        'ENAME':{
            '$regex':'S'
        }
    }
)
pd.DataFrame(result)

#'$regex': regex 뒤의 부분 조건을 만족하는 데이터를 출력한다.

result = db.emp.find(
    {'ENAME':{
        '$regex':'S'
        }
    }
)
pd.DataFrame(result)
#'$regex':'s' -> 어느 부분이든 's'를 포함하는 데이터 출력
# SELECT * FROM emp WHERE ENAME LIKE %s%

########################################################
result = db.emp.find(
    {'ENAME':{
        '$regex':'^S'
        }
    }
)
pd.DataFrame(result)
#'$regex':'^s' -> 맨 앞이 's'인 데이터 출력
# SELECT * FROM emp WHERE ENAME LIKE s%

#########################################################

result = db.emp.find(
    {'ENAME':{
        '$regex':'S$'
            }
    }
)
pd.DataFrame(result)
#'$regex':'s$' -> 맨 뒤가 's'인 데이터 출력
# SELECT * FROM emp WHERE ENAME LIKE %s
```

```python
##다음 sql 구문을 mongodb 형식으로 구동하라
# SELECT ENAME, JOB, SAL, DEPTNO FROM emp
# WHERE SAL >1250 AND SAL<=3000 
# AND JOB = 'SALESMAN'
# ORDER BY ENAME

result = db.emp.find(
    {
        'SAL':{'$gt':1250, '$lte':3000},
        'JOB':'SALESMAN'
    },
    {'ENAME':1,
     'JOB':1,
     'DEPTNO':1,
     '_id':0}
).sort('ENAME')

pd.DataFrame(result)
```

```python
#'ENAME'이 ALLEN인 데이터를 추출하라
result = db.emp.find(
    {'ENAME':'ALLEN'}
)
pd.DataFrame(result)
```

```python
# update({조건},{변경내용},{추가옵션}): 데이터의 수정


# update_one(): 특정 조건을 만족하는 데이터를 제일 앞 1개를 수정한다.
db.emp.update_one(
    {'ENAME':'FORD'},
    {
        '$set':{
            'SAL':880
        }
    }
)

pd.DataFrame(db.emp.find({},{'ENAME':1,'SAL':1}))
#db.emp.update_one(
#       {value 조건},
#        {'$set':{
#                 column 지정: 변경 데이터
#                }
#         }
#         )
# -> db에 있는 emp collection에 대해서 1개의 데이터를 수정하라.
# 이때, 바꿀 값은 value 조건을 만족하고 이때, 이 데이터의 
# column 지정값에 정의된 데이터를 변경 데이터로 변경한다.




#############################3
################################ 그냥 내장 함수에서 복수의 변경값들을 
# FOR 문을 통해 리스트로 만들어서 변경한다고 생각하자

db.emp.update_many(
                {'JOB':'SALESMAN'},
                [
                    {'$set':
                        {
                        'SAL':
                            {
                                '$multiply': [1.5,"$SAL"]
                            }
                        }
                    }
                ]
            )

#db.emp.update_many(
#           {value 조건},
#              {'$set':{
#                       columns 지정: 변경 데이터
#                       }
#               }    
#                   )
pd.DataFrame(db.emp.find())
```

```python
#delete: 데이터를 삭제하는 명령어
#        실제 몽고 내에서는 remove 명령어를 사용하고 
#        파이썬 적용시 delete 명령어를 사용한다.

db.emp.delete_one(
                    {
                        'JOB':'PRESIDENT'
                    }
                 )

#db.collenction.delete_one({조건}): 조건을 만족하는 db 내에 지정한 
#                                   collection의 데이터 중
#                                   가장 먼저 나온 데이터를 삭제한다         
```
