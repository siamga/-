## 분류 알고리즘
### 의미
* 학습데이터로 주어진 데이터의 피처와 레이블 값을 알고리즘으로 학습해 모델을 만들고 여기에 새로운 데이터가 주어질 경우 미지의 데이블 값을 예측하는 것

### 종류
1. 베이즈 통계(Bayes)와 나이브 베이즈(Naive Bayes)
2. 로지스틱 회귀(Logistic Regression)
3. 결정트리(Decision Tree)
4. 서포트 베거 머신(svm= Support Vector Machine)
5. 최소 근접 알고리즘(k-nn, Nearest Neighbor)
6. 심층 신경망 알고리즘(dnn, Neural Network)
7. 서로 다른 알고리즘의 결합(앙상블, Ensemble)


### 결정트리와 앙상블
#### 결정트리
##### 의미
* 데이터에 있는 규칙을 학습을 통해 찾아내 Tree기반의 분류 규칙을 만들어낸다.(if-Else기반 규칙)
<br>= 분할기준: 분할점
#### 장단점
* 장점
    -매우 쉽고 유연한 적용 가능
    - 스케일링이나 정규화 등의 사전 가공 영향이 매우 적다
* 단점
    - 예측 성능 향상을 위한 복잡한 구조가 과적합 문제를 가져올 수 있다.
    - 이런 단점이 앙상블 기법에서 다양한 알고리즘을 결합해 예측 성능을 향상시킬 수 있는데, 이때, 결정트리 모형이 사용된다.
    <br>ex:GBM,XGBoost,LightGBM
#### 용어
* 노드(nod): 데이터의 지점
* depth: 분할 깊이
* 최종적인 분할 개수 n개를 기준으로 n분할이라 부른다.

##### 정보 균일도 측정 방법
###### 정보이득(information Gain)
1. 엔트로피(entropy)
    * 주어진 데이터 집합의 혼잡도를 의미한다.
    (ex: 서로 다른 값이 섞임-> 엔트로피가 높음
     <br>같은 값이 섞임 -> 엔트로피가 낮음)

2. 정보이득지수
    * 1 - (엔트로피 지수)......(1)
    <br>=(1번째 노드 엔트로피 지수) - (2번째 노드 엔트로피 지수)......(2)
    * (1): 해당기준을 통해 얼마나 잘 분류되었는지를 의미
    * (2): 두번째 분류가 첫번째 분류에 비해 얼마나 잘 분류했는지를 의미

3. 지니계수
    * 주어진 데이터의 혼잡도를 0~1까지의 수로 표현한 수
    * 1로 갈수록 균일도가 낮아져서 분할 기준 설정 시, 지니계수가 낮아지는 기준을 설정하여 분할한다.
    * 분할 대상은 균일도가 낮은 1에 가까운 대상을 선정하여 분할한다.

4. 비용함수
    * 노드에서 분할점(분할기준)을 선택하기 이전에 노드를 적용할 변수(분할 변수)를 선택할 때 사용하는 함수로 이 결과값이 작은 변수를 사용한다. 이것을 구하는 방식은 다양하지만 지금은 개념만 이해하자. 물론 이 적용 변수 선정에 따라서 지니계수나 엔트로피 같은 값들은 당연히 달라지겠지만 어디까지나 비용함수가 가장 작아지는 변수를 선택한다고만 이해하자

#### DecisionTree 시각화

```python
# DecisionTree 시각화
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')

dt_clf = DecisionTreeClassifier(random_state=156)

iris_data = load_iris()
X_train, X_test,y_train,y_test = train_test_split(iris_data.data,iris_data.target,test_size=0.2,random_state=11)

dt_clf.fit(X_train,y_train)

from sklearn.tree import export_graphviz

export_graphviz(dt_clf,out_file='tree.dot',class_names=iris_data.target_names,
               feature_names = iris_data.feature_names, impurity=True,filled=True)
#export_graphviz(): 결정트리를 dot 형식의 파일로 내보내는 method

import graphviz
with open('tree.dot') as f:
    dot_graph = f.read()
graphviz.Source(dot_graph)
```

## 앙상블(Ensemble)
### 유형과 특징
#### 유형
1) 보팅(voting)
* 여러 예측 모델을 직접 돌려서 가장 적절한 모델을 선정하는 방식(hard voting)
* 보팅 결과값의 평균화(soft voting)
2) 배깅(Bagging)
* 복원 추출 결과의 평균화
3) 부스팅(Boosting)
* 순차적으로 모델을 적용해서 약한 알고리즘에 가중치를 주어서 다시 모델에 적용하는 알고리즘 유형
#### 특징 
* 단일 모델의 약점을 다수의 모델로 결합하여 보완
* 결정 트리의 단점인 과적합을 분류기 결합 방식인 앙상블로 보완하고 직관적 분류 기준이라는 장점이 강화된다.
<br>-> 그래도 과적합 문제의 해소는 아니라는 점 주의하자

### 보팅(Voting)과 배깅(Bagging)
#### 의미
* 여러개의 분류기가 투표를 통해 최종 예측을 결정하는 방식
#### 차이점
* 보팅(Voting): 서로 다른 분류기의 결합
* 배깅(Bagging): 같은 알고리즘에 대한 데이터 샘플링을 다르게 수행하면서 보팅을 수행하는 것
#### Voting 예시
* sklearn의 VotingClassifier를 통한 예시

```python
import pandas as pd

from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

cancer = load_breast_cancer()

data_df = pd.DataFrame(cancer.data,columns=cancer.feature_names)
data_df.head()
```

```python
# 보팅에 적용할 개별 모델 선정: 로지스틱회귀, KNN
lr_clf = LogisticRegression()
knn_clf = KNeighborsClassifier()

# 개별 모델로 소프트 보팅 기반 앙상블 모델 구현
vo_clf = VotingClassifier(estimators=[('LR',lr_clf),('KNN',knn_clf)],
                         voting='soft')

X_train,X_test,y_train,y_test = train_test_split(cancer.data,cancer.target,test_size=0.2,random_state=156)
# Voting분류기를 통한 학습/예측/평가
vo_clf.fit(X_train,y_train)
pred = vo_clf.predict(X_test)
print(accuracy_score(y_test,pred))
print(vo_clf.score(X_test,y_test))

# 개별 모델의 학습/예측/평가
classifiers = [lr_clf,knn_clf]
for classifier in classifiers:
    classifier.fit(X_train,y_train)
    pred = classifier.predict(X_test)
    class_name = classifier.__class__.__name__
    print(class_name,accuracy_score(y_test,pred))
```

### 배깅(Bagging)
* 대표 예시
    - 랜덤 포레스트 방식을 이용해 전체 데이터를 배깅 방식으로 샘플링을 한 이후 개별 학습을 수행하여 최종적으로 모든 분류기에 대해 보팅을 통해 예측 결정을 진행한다.
    

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X, y = make_moons(n_samples=500, noise=0.30, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)\

bag_clf = BaggingClassifier(DecisionTreeClassifier(), n_estimators=500, max_samples=100, bootstrap=True, n_jobs=-1,verbose=3)

bag_clf.fit(X_train, y_train)
y_pred = bag_clf.predict(X_test)

accuracy_score(y_test, y_pred)
```

### 부스팅(Boosting)
#### 의미
* 여러개의 약한 학습기를 순차적으로 학습-예측을 하면서 틀린 예측에 대해 가중치를 부여하면서 오류를 개선하면서 학습하는 방식
    <br>(ex: AdaBoost,GradientBoost)
* 수행시간이 엄청 걸린다
#### GBM(Gradient Boost Machine)
* 모델 예측치에 대한 잔차를 반복적으로 분류기에 넣어서 오류를 줄이는 방식의 기법
* 오류값 = 실제값 - 예측값
#### 서포트 벡터 머신(SVM,Support Vector Machine)
* 분류와 회귀 방식에 모두 사용
* 선형, 비선형 모두 적용 가능
* 비확률적 이진 선형 분류 모델 적용 가능

```python
from sklearn import datasets
iris = datasets.load_iris()
X = iris.data
y = iris.target
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=222)

from sklearn import svm
clf = svm.SVC(gamma=0.001,C=100)
clf.fit(X_train,y_train)
clf.score(X_test,y_test)

# gamma: 결정 경계의 복잡성, 높을수록 인식률은
#        상승하지만 과도한 연산이 수반된다.
# C : 잘못된 값 분류 시 부여하는 벌점
# 큰값: 오버피팅, 작은값: 언더피팅 문제 발생
```

### KNN알고리즘
#### 의미
* 특정 지점을 기준으로 거리에 따라서 분류를 하는 알고리즘

#### 예시
* 분류기준: a, b, k=5
    - 새로운 데이터 z에 대해서 가장 가까운 k(=5)개의 데이터의 분포
    > a: 3개, b: 2개
    - 따라서 새로운 데이터 z는 a로 분류
        
#### 장,단점
1) 장점
    * 학습이나 준비 시간이 필요 없이 신규 데이터 입력과 동시에 판단을 한다.

2) 단점
    * 특징 공간 내의 모든 정보가 필요하다.
    <br>(위의 예시의 경우 3 거리 내의 모든 데이터 정보 필요)

#### KNN 예시

```python
from sklearn.datasets import load_iris
iris = load_iris()
# print(iris.data)

from sklearn.model_selection import train_test_split

X = iris.data
y = iris.target

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=159)

print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

knn = KNeighborsClassifier(n_neighbors=6)
knn.fit(X_train,y_train)

y_pred = knn.predict(X_test)
score = metrics.accuracy_score(y_test,y_pred)
score
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X, y)
#0 = setosa, 1=versicolor, 2=virginica
classes = {0:'setosa',1:'versicolor',2:'virginica'}
# 아직 보지 못한 새로운 데이터를 제시해보자.
x_new = [[3,4,5,2],
[5,4,2,2]]
y_predict = knn.predict(x_new)
print(classes[y_predict[0]])
print(classes[y_predict[1]])
```
